Cost function (mean squred error)

```python
def cost_function(X, Y, W):
    """
    Calculates Mean Squared Error cost.
    """
    n = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

# Gradient Descent Implementation

```python
def gradient_descent(X, Y, W, alpha, iterations):
    """
    Performs gradient descent for linear regression.
    """
    cost_history = [0] * iterations
    m = len(Y)

    for i in range(iterations):
        # Step 1: Prediction
        Y_pred = np.dot(X, W)

        # Step 2: Loss
        loss = Y_pred - Y

        # Step 3: Gradient
        dw = (1 / m) * np.dot(X.T, loss)

        # Step 4: Update weights
        W = W - alpha * dw

        # Step 5: Cost
        cost_history[i] = cost_function(X, Y, W)

    return W, cost_history
```

RMSE Implementation

```python
def rmse(Y, Y_pred):
    """
    Calculates Root Mean Squared Error.
    """
    rmse = np.sqrt(np.mean((Y - Y_pred) ** 2))
    return rmse
```

R-Squared Implementation

```python
def r2(Y, Y_pred):
    """
    Calculates R-squared score.
    """
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    r2_score = 1 - (ss_res / ss_tot)
    return r2_score
```

Main Function

```python
def main():
    # Load dataset
    data = pd.read_csv("/content/drive/MyDrive/AI_COURSE_5/student.csv")

    # Features and Target
    X = data[['Math', 'Reading']].values
    Y = data['Writing'].values

    # Train-test split
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Initialize parameters
    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Train model
    W_optimal, cost_history = gradient_descent(
        X_train, Y_train, W, alpha, iterations
    )

    # Predictions
    Y_pred = np.dot(X_test, W_optimal)

    # Evaluation
    model_rmse = rmse(Y_test, Y_pred)
    model_r2 = r2(Y_test, Y_pred)

    # Results
    print("Final Weights:", W_optimal)
    print("Cost History (First 10):", cost_history[:10])
    print("RMSE:", model_rmse)
    print("R-Squared:", model_r2)

if __name__ == "__main__":
    main()
```
